```python
# Import libraries.
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load data.
df_cars = pd.read_csv('Cars93.csv', header=0)
df_cars.head()
```

### 1. Boxplot

```python
# Boxplot for revs per mile.
# Selecting Audi, Hyundai, Suzuki and Toyota manufacturers.
manufacturers_to_select = ['Audi', 'Hyundai', 'Suzuki', 'Toyota']
df_selected_manufacturers = df_cars[df_cars['Manufacturer'].isin(
manufacturers_to_select
)]
# Ploting the box plot.
plt.figure(figsize=[10,6])
sns.boxplot(
    data=df_selected_manufacturers, x='Manufacturer', y='Rev.per.mile', 
    hue='Manufacturer', showmeans=True, legend=True
)
plt.title('Rev Per Mile by Manufacturer')
plt.show()
```

**Boxplot question:**  
Which of these manufacturers has the car with the highest revs per mile?


**Boxplot answer:**


Among the Audi, Hyundai, Suzuki and Toyota, manufacturer of the Suzuki 
has the car with highest revs per mile.

### 2. Histogram

```python
# Histogram of MPG.
plt.figure(figsize=[10, 6])
sns.histplot(x=df_cars['MPG.city'],  bins=10, label='MPG.city')
sns.histplot(x=df_cars['MPG.highway'], bins=10,  label='MPG.highway')

plt.xlabel('Miles per Gallon')
plt.ylabel('Frequency')
plt.title('Histogram MPG in the City and on the Highway')
plt.legend()
plt.show()
```

**Histogram question:**  
Is it generally more fuel efficient to drive in the city or on the highway?

**Histogram answer:**

On the highway, It is more fuel efficient to drive than in the city.

### 3. Lineplot

```python
# Lineplot (Wheelbase vs turning circle).
plt.figure(figsize=[10, 6])
sns.lineplot(data=df_cars, x='Wheelbase', y='Turn.circle')
plt.title('Wheelbase Vs Turning Circle')
plt.show()
```

**Lineplot question:**  
What is this relationship? What happens when the wheelbase gets larger?


**Lineplot answer:**  


This is a Non linear relashionship. 
Overall, When wheelbase gets larger turn circle increases.

### 4. Barplot

```python
# Barplot for mean horsepower.
df_groupby_type = df_cars.groupby(by=['Type'])
# Getting the mean for each type across all columns.
df_mean_horsepower = round(df_groupby_type.mean(numeric_only=True), 0) 

# Mean horsepower for each type of vehicle.
plt.figure(figsize=[10, 6])
sns.barplot(x='Type', y='Horsepower', data=df_mean_horsepower, hue='Type')
plt.title('Mean Horsepower Vs Type')
plt.show()
```

**Barplot question:**  
Does a larger car mean more horsepower?



**Barplot answer:**

Yes, Larger car mean more horsepower.
