```python
# Import pandas.
import pandas as pd
```

```python
# Create a DataFrame with balance.txt.
df = pd.read_csv('balance.txt', sep=" ")
df.head()
```

```python
# Comparing average income based on ethnicity.
print("\nAverage Income based on Ethnicity")
df.groupby('Ethnicity')['Income'].mean()
```

```python
# Higher balance according to marital status.
print("\nHigher balance  according to marital status")
df.groupby('Married')['Balance'].mean().sort_values(ascending=False)
```

```python
# Highest income of the data set.
print("\nHighest income:", df['Income'].max())
```

```python
# Lowest income of the data set.
print("\nLowest income:", df['Income'].min())
```

```python
# Sum of the cards.
print("\nSum of cards:", df['Cards'].sum())
```

```python
# Female Vs Male count.
print("Female Vs Male count")
df.groupby('Gender')['Gender'].count()
```
